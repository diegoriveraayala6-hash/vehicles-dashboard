```python
import pandas as pd
import plotly_express as px

car_data = pd.read_csv('../vehicles_us.csv')

car_data.head()
```

```python
fig2 = px.scatter(car_data, x='odometer', y='price', title='Precio vs Odómetro')
fig2.show()
```

```python
fig = px.histogram(car_data, x='odometer', title='Distribución del Odómetro')
fig.show()
```
